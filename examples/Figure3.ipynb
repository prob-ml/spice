```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lineplot(x = range(12,60,12), y=[1.88,3.64,5.39,6.83], marker = 'o')
sns.lineplot(x = range(12,60,12), y=[24.14,45.23,65.16,80.97], marker = 'o')
sns.lineplot(x = range(12,60,12), y=[2.10,2.69,3.43,4.19], marker = 'o')
sns.lineplot(x = range(6,36,6), y=[65.85,194.81,407.33,711.07,1611.61], marker = 'o')
plt.yscale('log')
plt.title("Training Time for Merfish Slices")
plt.xlabel("Number of Training Slices")
plt.ylabel("Log Time")
plt.legend(labels=["LightGBM", "deepST", "XGBoost", "MESSI"], loc="upper right")
plt.show()
```
